## Árbol de Decisión CART con Regresión. ¿Gripe o resfriado?

##### Objetivo. Predecir si una persona tendrá gripe o resfriado

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
```

## Creación del Dataset

```python
##Dataset
data = {
    'Fiebre_Alta': [1, 1, 0, 0, 1],
    'Dolor_Muscular': [1, 1, 0, 1, 0],
    'Congestion_Nasal': [0, 1, 1, 0, 1],
    'Duracion': [7, 10, 3, 5, 8] # Variable objetivo numérica
}
```

```python
df = pd.DataFrame(data)
#Se divide en dos parte: Característica(X) y variables objetivo(y)
X = df.drop('Duracion', axis=1) # Características . drop elimina la columna Duracion, axis = 1 indica que es una coluna
y = df['Duracion'] # Target variable, la columna Duracion

print("\nDataset Original: \n")
df
```

## Entrenar el modelo

```python
model = DecisionTreeRegressor(
    criterion='squared_error', #Usa MSE para regresión
    max_depth=2,  #Limitamos la profundidad para simplificar
    random_state=42
)
model.fit(X, y)
```

## Visualizar el árbol

```python
plt.figure(figsize=(12, 6))
plot_tree(
    model,
    feature_names = X.columns,
    filled=True,
    rounded=True,
    precision=1
)
plt.title("Árbol de Decisión - Clasificación de Enfermedades(Gripe ó Resfriado)")
plt.show()
```

### Predicciones

```python
# Predecir un nuevo paciente: Fiebre = Si(1), Dolor Muscular = No(0), Congestión Nasal = Si(1)
nuevo_paciente = pd.DataFrame([[1, 0, 1]], columns=X.columns)
prediccion = model.predict(nuevo_paciente)
print(f"\nDuración predicha: {prediccion[0]:.1f} días\n")
```
