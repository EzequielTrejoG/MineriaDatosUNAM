## Árbol de Decisión CART. ¿Gripe o resfriado?

##### Objetivo. Predecir si una persona tendrá gripe o resfriado

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

## Creación del Dataset

```python
##Dataset
data = {
    'Fiebre_Alta': [1, 1, 0, 0, 1],
    'Dolor_Muscular': [1, 1, 0, 1, 0],
    'Congestion_Nasal': [0, 1, 1, 0, 1],
    'Diagnostico': [1, 1, 0, 0, 1] # 1 = Gripe 0 = Resfriado
}
```

```python
df = pd.DataFrame(data)
#Se divide en dos parte: Característica(X) y variables objetivo(y)
X = df.drop('Diagnostico', axis=1) # Características . drop elimina la columna Diagnostico, axis = 1 indica que es una coluna
y = df['Diagnostico'] # Target variable, la columna Diagnostico

print("\nDataset Original: \n")
df
```

## Entrenar el modelo

```python
model = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
model.fit(X, y)
```

## Visualizar el árbol

```python
plt.figure(figsize=(8, 4))
plot_tree(
    model,
    feature_names = X.columns,
    class_names = ['Resfriado', 'Gripe'],
    filled=True,
    rounded=True
)
plt.title("Árbol de Decisión - Clasificación de Enfermedades(Gripe ó Resfriado)")
plt.show()
```

### Predicciones

```python
# Predecir un nuevo paciente: Fiebre = No(0), Dolor Muscular = No(0), Congestión Nasal = Si(1)
nuevo_paciente = pd.DataFrame([[0, 0, 1]], columns=X.columns)
prediccion = model.predict(nuevo_paciente)
print("\nPREDDICIONES: \nNuevo Paciente: \n")
print(f"Síntomas: \n{nuevo_paciente}")
if prediccion[0] == 1:
    print("\nEl paciente tiene Gripe.")
else:
    print("\nEl paciente tiene Resfriado.")
```
